-->If stopwords are not removed, the dataset will contain many common words like “the”, “is”, and “and”. These words appear very often but usually do not carry important meaning. Because of this, the model may take longer to process the text and important keywords may become harder to identify. 

3. Provide two real-world scenarios where removing stopwords can be harmful.

-->1)In sentiment analysis, some stopwords like “not” are very important. For example, the sentence “I do not like this movie” expresses a negative opinion. If the word “not” is removed, the sentence becomes “I like this movie”, which completely changes the meaning.
2)In search engines or question answering systems, stopwords can help understand the full meaning of a question. For example, a phrase like “to be or not to be” would lose its meaning if the stopwords were removed.

4. Explain the difference between stemming and lemmatization.

-->1)Stemming is a simple technique that removes word endings using basic rules. Sometimes the result is not a real word. For example, the word “studies” may become “studi”.
2)Lemmatization is more advanced because it looks at the meaning of the word and converts it into its correct dictionary form. For example, “running” becomes “run” and “better” becomes “good”.

## Task 2:Build Advanced Preprocessing Function

````python
import re
import nltk
from nltk.tokenize import word_tokenize

# run once if tokenizer not installed
# nltk.download('punkt')

def preprocess_text(text):
    
    # handle empty input
    if not text:
        return [], ""
    
    # convert to lowercase
    text = text.lower()
    
    # remove urls
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # remove email patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # remove numbers
    text = re.sub(r'\d+', '', text)
    
    # reduce repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # tokenize using NLTK
    tokens = word_tokenize(text)
    
    cleaned_tokens = []
    
    for token in tokens:
        if token.isalpha():
            if len(token) > 2 or token in ["no", "not"]:
                cleaned_tokens.append(token)
    
    cleaned_sentence = " ".join(cleaned_tokens)
    
    return cleaned_tokens, cleaned_sentence
````

## Task 3: Stress Testing


````python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens, cleaned = preprocess_text(s)
    
    print("Original Text:", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 40)
````

## Task 4: Token Analytics

````python
import statistics

for s in sentences:
    tokens, cleaned = preprocess_text(s)
    
    if len(tokens) == 0:
        avg_len = 0
    else:
        avg_len = sum(len(t) for t in tokens) / len(tokens)
    
    print("Sentence:", cleaned)
    print("Total tokens:", len(tokens))
    print("Unique tokens:", len(set(tokens)))
    print("Average token length:", round(avg_len,2))
    print("-"*40)
````

## Task 5: Frequency Analysis

````python
from collections import Counter

all_tokens = []

for s in sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
````

## Task 6: Build Full Pipeline

````python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
````
